```python
# Liam Fruzyna
# COSC 4610
# Assignment 3
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic = titanic.drop('Name', 1)
titanic['Male'] = (titanic['Sex'] == 'male').astype(int)
titanic = titanic.drop('Sex', 1)

titanic.head()
```

```python
frequent_itemsets = apriori(titanic, min_support=0.3, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```
