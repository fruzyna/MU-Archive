```python
import pandas as pd
```

```python
df = pd.read_table('house_selling_prices.txt')
df.head()
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
%matplotlib inline

# make a scatterplot
sns.regplot(x=df['Price'], y=df['Size'])
```

```python
# make a distribution plot
sns.distplot(df['Price'])
```

```python
# Somewhat normal
# Test for normallity using the Kolmogorov-Smirnov test
from scipy.stats import kstest

test_stat = kstest(df['Price'], 'norm')
test_stat
```

```python
# Selling price according to bedrooms
sns.regplot(x=df['Price'], y=df['Bedrooms'])
```

```python
# Selling price according to bathrooms
sns.regplot(x=df['Price'], y=df['Baths'])
```

```python
# Bedrooms are more valuable than bathrooms

# Compare number of bedrooms and bathrooms
sns.regplot(x=df['Bedrooms'], y=df['Baths'])
```

```python
import matplotlib.pyplot as plt

# Show price distribution by number of bedrooms
multiples = sns.FacetGrid(df, col='Bedrooms')
multiples = multiples.map(plt.hist, 'Price')
```

```python
# Show price distribution by number of bathrooms
multiples = sns.FacetGrid(df, col='Baths')
multiples = multiples.map(plt.hist, 'Price')
```

```python
# Most houses in the city are 3 bedroom 2 bath

# Show distribution of bathrooms by bedrooms count
multiples = sns.FacetGrid(df, col='Bedrooms')
multiples = multiples.map(plt.hist, 'Baths')
```
