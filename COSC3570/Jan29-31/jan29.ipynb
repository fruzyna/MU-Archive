```python
import pandas as pd

artwork = pd.read_csv('artwork_data.csv', dtype='unicode', index_col='id') #load in from file
artwork = artwork.drop(['thumbnailCopyright', 'thumbnailUrl', 'url'], axis=1) #drop unnessesary columns
artwork = artwork.drop([1035], axis=0) #drop unneeded rows
artwork.head()
```

```python
artwork.describe()
```

```python
#1: How many distinct artists are there?
len(artwork['artist'].unique())
```

```python
#2: How many artworks are by William Blake?
len(artwork[artwork['artist'] == 'Blake, William'])
```

```python
#3: What is the artwork with the largest area?
artwork['width'] = pd.to_numeric(artwork['width'], errors='coerce')
artwork['height'] = pd.to_numeric(artwork['height'], errors='coerce')
area = artwork['width'].multiply(artwork['height'])
artwork['area'] = area
max = area.max()
artwork[area == max]
```
