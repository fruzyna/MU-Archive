```python
import pandas as pa
import datetime as dt

bikes = pa.read_csv('bikes.csv', ';', encoding = 'latin-1', dayfirst = True, index_col='Date', parse_dates = True)
day_week = bikes.index.weekday_name
bikes['Day of Week'] = day_week
days = bikes.groupby('Day of Week')
means = days.mean()
means.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
means
```

```python
sums = days.sum()
sums
```

```python
%matplotlib inline
means.plot(kind="bar", figsize=(20, 10))
```
