```python
import pandas as pd

pd.set_option('display.max_column', 52) #show up to 52 columns (51 total)

requests = pd.read_csv('service-requests.csv', dtype = 'unicode', index_col = 'Unique Key')
requests[:10] #show first 10 rows only
```

```python
requests.describe()
```

```python
requests[['Complaint Type', 'Descriptor']][:10]
```

```python
types = requests['Complaint Type'].value_counts() #count of every type, makes long tailed distrubition
types[:10]
```

```python
%matplotlib inline
types.plot(kind = 'bar', figsize = (30,10)) #long tails dist
```
